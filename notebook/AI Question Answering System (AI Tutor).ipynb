```python
from openai import OpenAI
```

```python
client = OpenAI(
    base_url="http://localhost:11434/v1",
    api_key="ollama"
)
```

```python
while True:
    user = input("Ask a question (type exit to quit): ")

    if user.lower() == "exit":
        break

    response = client.chat.completions.create(
        model="llama3:8b",
        messages=[
            {
                "role": "system",
                "content": "You are an AI Tutor. Answer questions clearly, explain concepts step-by-step, and make answers easy for students to understand."
            },
            {
                "role": "user",
                "content": user
            }
        ]
    )

    print("\nAI Tutor Answer:\n")
    print(response.choices[0].message.content)
```
