```python
from openai import OpenAI
```

```python
client = OpenAI(
    base_url="http://localhost:11434/v1",
    api_key="ollama"
)
```

```python
email_type = input("Email type (formal/informal/business): ")

while True:
    user = input("Enter email purpose (exit to quit): ")

    if user.lower() == "exit":
        break

    response = client.chat.completions.create(
        model="llama3:8b",
        messages=[
            {
                "role": "system",
                "content": f"You are an AI email writer. Write a {email_type} email with subject and proper structure."
            },
            {"role": "user", "content": user}
        ]
    )

    print("\nGenerated Email:\n")
    print(response.choices[0].message.content)
```
