```python
from openai import OpenAI
```

```python
client = OpenAI(
    base_url="http://localhost:11434/v1",
    api_key="ollama"
)
```

```python
while True:
    user=input("Enter something to generate a code: ")
    if user.lower()=="exit":
        break
    response = client.chat.completions.create(model="llama3:8b",messages=[{"role":"system","content":"You are a programer"},{"role": "user", "content": user}])
    print(response.choices[0].message.content)
```
