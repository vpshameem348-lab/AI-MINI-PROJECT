```python
from openai import OpenAI
```

```python
client = OpenAI(
    base_url="http://localhost:11434/v1",
    api_key="ollama"
)
```

```python
while True:
    user=input("Enter something to text summarize: ")
    if user.lower()=="exit":
        break
    response = client.chat.completions.create(model="llama3:8b",messages=[{"role":"system","content":"You are an AI text summarizer.Summarize long content into a short and clear summary with key points.."},{"role": "user", "content": user}])
    print(response.choices[0].message.content)
```
