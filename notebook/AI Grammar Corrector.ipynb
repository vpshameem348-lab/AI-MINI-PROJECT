```python
from openai import OpenAI
```

```python
client = OpenAI(
    base_url="http://localhost:11434/v1",
    api_key="ollama"
)
```

```python

while True:
    user = input("Enter sentence (exit to quit): ")

    if user.lower() == "exit":
        break

    response = client.chat.completions.create(
        model="llama3:8b",
        messages=[
            {
                "role": "system",
                "content": "You are an English grammar teacher. Correct the sentence and explain the correction."
            },
            {"role": "user", "content": user}
        ]
    )

    print("\nGrammar Correction:\n")
    print(response.choices[0].message.content)
```
